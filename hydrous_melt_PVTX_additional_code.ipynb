```python
import numpy as np

'''
Compositions from table 1 in Jing and Karato 2012
Values for f calculated using this script are given below the compostions
Compared to Jing and Karato (table 8) in brackets
'''
''' s3 '''
#Xw_SiO2=36.8; Xw_Al2O3=3.4; Xw_FeO=16; Xw_MgO=30; Xw_CaO=13.8; Xw_H2O=0
''' f = 0.542794039455279 (0.53) '''
''' s7 '''
#Xw_SiO2=36.1; Xw_Al2O3=3.3; Xw_FeO=20.9; Xw_MgO=27.1; Xw_CaO=12.68; Xw_H2O=0
''' f = 0.5434751208341843 (0.54) '''
''' s6 '''
#Xw_SiO2=35.2; Xw_Al2O3=3.3; Xw_FeO=26.8; Xw_MgO=23.7; Xw_CaO=11; Xw_H2O=0
''' f = 0.543990147318818 (0.53) '''
''' s7_2 '''
#Xw_SiO2=35.4; Xw_Al2O3=3.3; Xw_FeO=20.5; Xw_MgO=26.5; Xw_CaO=12.3; Xw_H2O=2.0
''' f = 0.5120575463331519 (0.51) '''
''' s6_3 '''
#Xw_SiO2=34.2; Xw_Al2O3=3.2; Xw_FeO=26; Xw_MgO=23; Xw_CaO=10.7; Xw_H2O=3.0
''' f = 0.49719822322477536 (0.49) '''
''' s6_5 '''
#Xw_SiO2=33.5; Xw_Al2O3=3.1; Xw_FeO=25.5; Xw_MgO=22.5; Xw_CaO=10.5; Xw_H2O=5.0
''' f = 0.4740291201782252 (0.46) '''
''' s6_7 '''
Xw = np.array([32.7, 3.0, 24.9, 22.1, 10.2, 7.0])
T = 2173
P = 14.3
V0 = 18.21E-6
''' f = 0.4623533281353758 (0.43) '''

# constants
N_A=6.02214086E23
T_ref = 1673
R = 8.31446  # ideal gas constant
mu = 4 / 3  # attractive exponent
nu = 8 / 3  # repulsive exponent

# Optimised hard sphere parameters from table 7 of Jing & Karato 2012 (diameter, temperature dependence, volume dependence)
sigma_0 = np.array([0.369, 0.326, 0.286, 0.272, 0.315, 0.172])
eta_i = np.array([-.04, -.03, 0.01, 0.04, -.06, .13])
zeta_i = np.array([.69, .58, .15, .14, .07, 0])

# molar masses
M_O = 15.9994
M = np.array([
    28.0855 + 2 * M_O, 2 * 26.9815 + 3 * M_O, 55.845 + M_O, 24.305 + M_O,
    40.078 + M_O, 2 * 1.0079 + M_O
])

# thermal expansivities of components for linear mixing model (cc mol-K)
# SiO2: Lange & Carmichael (1987), Table 9
# Al2O3: Lange & Carmichael (1987), Table 9
# H2O: Ochs & Lange (1997)
# FeO: Kress & Carmichael (1989)
# MgO: Lange (1996) 
# CaO: Lange (1996)
dVdT = np.array([0, 2.62E-9, 10.86E-9, 2.92E-9, 3.27E-9, 3.74E-9])

# partial molar volume of components at 1 atm for linear mixing model
# SiO2: Lange & Carmichael (1987), Table 9
# Al2O3: Lange & Carmichael (1987), Table 9
# H2O: Ochs & Lange (1997)
# FeO: 
# MgO: Lange (1996) 
# CaO: Lange (1996)
Vi = np.array([26.9E-6, 37.11E-6, 27.75E-6, 13.65E-6, 10.71E-6, 15.41E-6])

# convert to mole fraction
Xm = Xw / M
Xm_tot = np.sum(Xm)
X = Xm / Xm_tot

# Molar volume of the melt at experimental PT calculated from experimental density,
# ambient pressure density and ambient V0 in m^3 mol-1
V = (1 - (3.51 - 2.57) / 2.57) * V0
Va = np.sum(X*Vi)
display(V0,Va,V)

# Thermal expansivity calculated from linear mixing model
alpha_0 = np.sum((X*dVdT)/(X*Vi)) #1e-4
display(alpha_0)

# Sphere diameter for each component at elevated PT
sigma_i = (sigma_0 * ((T / T_ref)**eta_i)) * np.exp(
    (zeta_i / 3) * alpha_0 * (T - T_ref)) * ((V / V0)**(zeta_i / 3))

# Volume of a mole of spheres in nm^3 at 1 atm
Vm0 = (1/6)*np.pi*N_A* (np.sum(X * (sigma_0**3)))

# Volume of a mole of spheres in nm^3 at experimental PT
Vm = (1/6)*np.pi*N_A * (np.sum(X * (sigma_i**3)))

# Molar packing fraction at 1 atm; the same as the sum of fi0
f0 = (Vm0 * 1E-27) / V0

# Molar packing fraction at experimental PT; the same as the sum of fi
f = (Vm * 1E-27) / V

# Partial molar packing fraction at experimental PT (Eq. 23 in Jing & Karato, 2011)
fi = (1 / (6 * (V / 1E-27))) * np.pi * (sigma_i**3) * X * N_A

# Partial molar packing fraction at 1 atm (Eq. 23 in Jing & Karato, 2011)
fi0 = (1 / (6 * (V0 / 1E-27))) * np.pi * (sigma_0**3) * X * N_A

# Average deformability of a mole of spheres
zeta = (1 / Vm) * np.sum(X * Vm * zeta_i)

# Eq. 26 in Jing & Karato, 2011
def func_delta(fi, sigma, X):
    y = np.zeros((len(sigma), len(sigma)))
    for i in range(len(sigma)):
        for j in range(i + 1, len(sigma), 1):
            y[i, j] = ((fi[i] * fi[j]**(1 / 2)) * (((sigma[i] - sigma[j])**2) /
                                                   (sigma[i] * sigma[j])) *
                       ((X[i] * X[j])**(1 / 2)))
    return y


# Parameter representing the interaction of spheres of different sizes (Eq. 24 in Jing & Karato, 2011)
def func_y1(sigma, fi, X):
    y = np.zeros((len(sigma), len(sigma)))
    for i in range(len(sigma)):
        for j in range(i + 1, len(sigma), 1):
            y[i, j] = func_delta(fi, sigma, X)[i, j] * (sigma[i] + sigma[j]) * (
                (sigma[i] * sigma[j])**(-1 / 2))
    return np.sum(y)


# Parameter representing the interaction of spheres of different sizes (Eq. 25 in Jing & Karato, 2011)
def func_y2(sigma, fi, X, f):
    y = np.zeros((len(sigma), len(sigma)))
    for i in range(len(sigma)):
        for j in range(i + 1, len(sigma), 1):
            x = np.zeros(len(sigma))
            for k in range(len(sigma)):
                x[k] = (fi[k] / f) * ((
                    (sigma[i] * sigma_i[j])**(1 / 2)) / sigma[k])
            y[i, j] = func_delta(fi, sigma, X)[i, j] * np.sum(x)
    return np.sum(y)


# Describes the departure of the EoS from ideal mixing and is a function of packing fraction
def func_phi(sigma_i, fi, X, f):
    return (1 + ((1 - (3 * func_y1(sigma_i, fi, X))) * f) +
            ((1 - (3 * func_y2(sigma_i, fi, X, f)) * (f**2))) / ((1 - f)**3))

phi = func_phi(sigma_i,fi,X,f)
phi0 = func_phi(sigma_0,fi0,X,f0)

P = ((R*T)/V0)*((((1-zeta)*phi)-(phi0*((V0/V)**(mu-1))))+(zeta*phi0*((Vm0/Vm)**(nu-1))))/1e9
display(f0,P)
```
