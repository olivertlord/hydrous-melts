### Compositions studied

The table below defines all of the compositions studied in this work.

***

```python
# Load data

# Compositions
compositions = pd.read_csv('../hydrous-melt/PVTX data/hydrous-melts_compositions.csv', index_col=0)

compositions
```

```python
# Extract each compositional vector from Pandas DataFrame and convert to NumPy array
all_compositions = compositions.iloc[:,0]
En = compositions.filter(like='En', axis=0).to_numpy()
Fo = compositions.filter(like='Fo', axis=0).to_numpy()
MS = compositions.filter(like='MS', axis=0).to_numpy()
Br = compositions.filter(like='Br', axis=0).to_numpy()

# Create lists of ternary co-ordinates for all compositions along each vector
En_points,Fo_points,MS_points,Br_points = [], [], [], []
En_points.append((En[:,6], En[:,4], En[:,5]))
Fo_points.append((Fo[:,6], Fo[:,4], Fo[:,5]))
MS_points.append((MS[:,6], MS[:,4], MS[:,5]))
Br_points.append((Br[:,6], Br[:,4], Br[:,5]))
```

```python
# Create ternary plot for studied compositions

# Boundary and Gridlines
scale = 100
figure, tax = ternary.figure(scale=scale)

# Draw Boundary and Gridlines
tax.boundary(linewidth=1)
tax.gridlines(color="black", multiple=10)
tax.gridlines(color="black", multiple=10, linewidth=0.5)

# Set Axis labels and Title
tax.right_corner_label("$SiO_2$", fontsize=10, offset=-.1)
tax.top_corner_label("$H_2O$", fontsize=10, offset=0.25)
tax.left_corner_label("MgO", fontsize=10, offset=-.1)

# Set ticks
tax.ticks(axis='lbr', linewidth=0.5, multiple=10, offset=0.02, fontsize=6)

# Remove default Matplotlib Axes
tax.clear_matplotlib_ticks()
tax.get_axes().axis('off')

# Plot compositions
tax.scatter(En_points, marker='o', color='red', linewidths='0.5', alpha='1')
tax.scatter(Fo_points, marker='o', color='blue', linewidths='0.5', alpha='1')
tax.scatter(MS_points, marker='o', color='green', linewidths='0.5', alpha='1')
tax.scatter(Br_points, marker='o', color='orange', linewidths='0.5', alpha='1')

# Plot vectors
tax.line((50,0,50), (0,100,0), linewidth=1, color='red', linestyle="-")
tax.line((33.3,0,66.6), (0,100,0), linewidth=1, color='blue', linestyle="-")
tax.line((45,0,55), (0,100,0), linewidth=1, color='green', linestyle="-")

ternary.plt.show()
```
